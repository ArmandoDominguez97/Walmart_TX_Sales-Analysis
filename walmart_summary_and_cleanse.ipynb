```python
df = pd.read_csv('Walmart.csv', encoding_errors='ignore')

df.shape
```

```python
df.head
```

```python
df.describe()
#quantity column has a count of 10020, but all other columns have a count of 10051, which means there are 31 missing values in quantity column.
```

```python
df.info()
#unit_price and quantity columns have 31 missing values, which confirms the observation from describe() that there are 31 missing values in quantity column.
#unit_price column is a string object, which may need to be converted to a numeric data type for analysis.
```

```python
df.duplicated().sum()
#duplicated().sum() checks for duplicate rows in the dataset. In this case, there are 51 duplicate rows in the dataset.
```

```python
df.drop_duplicates(inplace=True)
#drop_duplicates() removes duplicate rows from the dataset.
df.duplicated().sum()
#after dropping duplicates, there are 0 duplicate rows in the dataset. this will help ensure that our analysis is not skewed by duplicate entries, but also affect the entire dataset.
```

```python
df.shape
```

```python
df.isnull().sum()
#isnull().sum() checks for missing values in each column of the dataset. In this dataset, there are 31 null values in both the unit_price and quantity columns, while all other columns have 0 null values.
```

```python
#dropping duplicates did not affect the number of missing values in the dataset.
#For this project we will drop rows with missing values in the interest of time, but we could and probably should explore other options for handling missing data, such as imputation or using algorithms that can handle missing values.
df.dropna(inplace=True)
#dropna() removes rows with missing values from the dataset.
```

```python
df.isnull().sum()
#double-checking that missing values have been removed. there are now 0 null values in the dataset, which means all rows with missing values have been dropped.
df.shape
#after dropping rows with missing values, the dataset has 9969 rows and 11 columns
```

```python
#recall that unit_price column is a string, which will need to be converted to a numeric data type for analysis.
df['unit_price'] = df['unit_price'].str.replace('$', '').astype(float)
#this code removes the dollar sign from the unit_price column and converts the values to float data.
```

```python
df.head()
#double-checking that values in unit_price column no longer have dollar signs.
```

```python
df.info()
#double-checking that unit_price column is now a float data type.
```

```python
#adding a total column to the dataset by multiplying unit_price and quantity columns together.
df['total'] = df['unit_price'] * df['quantity']
#this code creates a new column called total by multiplying the unit_price and quantity columns together. this will allow us to analyze total sales for each transaction.
```

```python
df.head()
#double-checking that total column has been added to the dataset and contains the correct values.
```
